### Spectrogram

```python
clip = TRAIN_CAT_DIR+"cat_78.wav"
```

```python
Audio(clip) # this is a clip of 7 distinct meows
```

```python
amplitude_series, sampling_rate = librosa.load(clip)
librosa.display.waveplot(amplitude_series, sampling_rate) # plot amplitude vs time
plt.show()
```

```python
D = librosa.stft(amplitude_series)
log_power = librosa.logamplitude(D**2, ref_power=np.max)
librosa.display.specshow(log_power,y_axis='linear', x_axis='time')
plt.colorbar()
plt.show()
```
